```python
df = pd.read_csv("~/code/states/state_areas.csv", header=[0, 1])
df.head(5)
```

```python
df_1 = df.copy()
df_1.head()
df_1 = df_1[:-3]
df_1.tail()
```

```python
state_area = df_1.iloc[:, df.columns.get_level_values(0)=='Land area[2]'] #[['Rank', 'sq mi', 'km²']]
# df_1.columns
state_area.columns = state_area.columns.droplevel(0)
state_area = state_area.rename(columns={'km²':'km^2'})
state_area['km^2'] = state_area['km^2'].str.replace(',', '')
state_area['km^2'] = state_area['km^2'].astype(int)
state_area.head()
```

```python
states = df_1.iloc[:, df.columns.get_level_values(1)=='State']['State']
states.head()
# states['State']
```

```python
state_area.set_index(states['State'], inplace=True)
state_area.drop(axis = 1, columns=['Rank', 'sq mi', '% land'], inplace=True)
```

```python
state_area.head()
```

```python
total_area = np.sum(state_area['km^2'])
total_area
```

```python
type(state_area['km^2'][0])
```

```python
state_area['prop'] = state_area['km^2'] / total_area
```

```python
state_area.head()
```

```python
state_area['pctl'] = state_area['prop'].cumsum()
# k = np.random.rand()
```

```python
def rand_state():
    k = np.argmax(state_area['pctl'] > np.random.rand())
    return df.iloc[k]
```
